## Descripción General del Dataset




Contiene información detallada sobre los viajes realizados por los Yellow Cabs (taxis amarillos en español) en la ciudad de Nueva York durante el año 2014. Este registro es recopilado por la Taxi and Limousine Commission (TLC).

En el dataset se identifica la logística del viaje, los puntos de origen y destino, y el desglose de los costos pagados por el pasajero. Cada fila representa un viaje individual completado, capturando datos desde el momento en que se activa el taxímetro hasta que se finaliza el servicio.

Está compuesto por las siguientes columnas:

* vendor_id: Código que indica el proveedor de tecnología que facilitó el registro del viaje (ej. VTS o CMT).

* pickup_datetime: Fecha y hora exacta en la que el pasajero subió al taxi y se inició el viaje.

* dropoff_datetime: Fecha y hora exacta en la que el pasajero bajó del vehículo y terminó el servicio.

* passenger_count: Número de pasajeros en el vehículo (dato introducido manualmente por el conductor).

* trip_distance: Distancia total del viaje recorrida en millas, reportada por el taxímetro del vehículo.

* pickup_longitude: Coordenada geográfica de longitud donde se inició el trayecto.

* pickup_latitude: Coordenada geográfica de latitud donde se inició el trayecto.

* rate_code: Código de la tarifa aplicada al viaje (ej. Tarifa estándar, JFK, Newark).

* store_and_fwd_flag: Bandera que indica si el registro se guardó en la memoria del vehículo antes de enviarse.

* store_and_fwd_flag: (Repetición del indicador de almacenamiento y reenvío de datos).

* dropoff_longitude: Coordenada geográfica de longitud donde finalizó el trayecto.

* dropoff_latitude: Coordenada geográfica de latitud donde finalizó el trayecto.

* payment_type: Método de pago utilizado por el cliente (ej. Tarjeta de crédito, Efectivo).

* fare_amount: Importe de la tarifa basado exclusivamente en el tiempo y la distancia calculados por el taxímetro.

* urcharge: Recargos adicionales aplicados (como cargos por hora pico o nocturnidad).

* mta_tax: Impuesto de $0.50 de la MTA aplicado automáticamente al viaje.

* tip_amount: Monto de la propina pagada por el cliente (registrada principalmente en pagos electrónicos).

* tolls_amount: Monto total pagado en peajes durante el trayecto.

* total_amount: Importe total cobrado al pasajero (incluye tarifa, propinas, impuestos y todos los recargos).

##Analisis

```python
import kagglehub
import pandas as pd
import os

#  Descargamos la última versión del dataset
path = kagglehub.dataset_download("kentonnlp/2014-new-york-city-taxi-trips")

print("Los archivos se descargaron en:", path)

#  Listamos los archivos para ver qué hay dentro
archivos = os.listdir(path)
print("\nArchivos disponibles para 'diagnosticar':")
for i, archivo in enumerate(archivos):
    print(f"{i+1}. {archivo}")

print("\n") # Imprime una línea en blanco para formateo
# The line below was attempting to print a filename that does not exist in the dataset.
# print("olist_customers_dataset.csv") # Imprimo el nombre de solo los valores unicos que son object que se estan analizando
print("\n") # Imprime una línea en blanco para formateo

# Construimos la ruta completa al archivo usando el 'path' que nos dio kagglehub
# Corrected the filename to 'nyc_taxi_data_2014.csv' which was found in the listed files.
archivo_objetivo = os.path.join(path, 'nyc_taxi_data_2014.csv')

# Lo cargamos en un DataFrame de Pandas
df_taxi = pd.read_csv(archivo_objetivo)

# Muestra una vista previa de las primeras filas del DataFrame df_taxi.
df_taxi.head()
```

```python


# Muestra un resumen técnico del DataFrame, incluyendo tipos de datos y valores no nulos.
print("\n\n\n1. Resumen técnico (df.info()):")
df_taxi.info()

# Muestra un resumen estadístico de las columnas numéricas del DataFrame.
print("\n\n\n2. Resumen estadístico:")
display(df_taxi.describe())

# Calcula y muestra el porcentaje de valores nulos para cada columna en el DataFrame.
print('\n\n\n3. Porcentaje de valores nulos por columna:')
display((df_taxi.isnull().sum() / len(df_taxi)) * 100)

print("\n\n\n4. Valores Nulos:")
# Calcula la cantidad de valores nulos por columna en el DataFrame.
display(df_taxi.isnull().sum())

# Calcula y muestra el número de filas duplicadas en el DataFrame
print(f"\n\n\n5. Filas Duplicadas: {df_taxi.duplicated().sum()}")
```

## Soluciones

###Rellenar valores faltantes

```python
# Reemplazar los valores nulos (NaN) en la columna 'review_comment_title' con la cadena 'no comento'.
df_taxi['store_and_fwd_flag'] = df_taxi['store_and_fwd_flag'].fillna('no aplica')
```

```python
# Calcula la mediana de la columna 'dropoff_longitude'
# La función .median() se utiliza para obtener el valor central de una serie numérica,
# siendo una medida de tendencia central menos sensible a valores atípicos que la media.
media_dropoff_longitude = df_taxi['dropoff_longitude'].median()

# Rellena los valores nulos (NaN) en la columna 'dropoff_longitude' con la mediana calculada.
# inplace=True modifica el DataFrame directamente sin necesidad de reasignar.
df_taxi['dropoff_longitude'].fillna(media_dropoff_longitude, inplace=True)

print(f"La mediana de 'dropoff_longitude' utilizada para rellenar es: {media_dropoff_longitude:.6f}")
print(f"Valores nulos restantes en 'dropoff_longitude': {df_taxi['dropoff_longitude'].isnull().sum()}")
```

```python
# Calcula la mediana de la columna 'dropoff_latitude'
# Al igual que con la longitud, la mediana es una medida robusta para valores centrales.
media_dropoff_latitude = df_taxi['dropoff_latitude'].median()

# Rellena los valores nulos (NaN) en la columna 'dropoff_latitude' con la mediana calculada.
# inplace=True aplica el cambio directamente al DataFrame.
df_taxi['dropoff_latitude'].fillna(media_dropoff_latitude, inplace=True)

print(f"La mediana de 'dropoff_latitude' utilizada para rellenar es: {media_dropoff_latitude:.6f}")
print(f"Valores nulos restantes en 'dropoff_latitude': {df_taxi['dropoff_latitude'].isnull().sum()}")
```

```python
# Calcula la mediana de la columna 'pickup_longitude'
media_pickup_longitude = df_taxi['pickup_longitude'].median()

# Rellena los valores nulos (NaN) en la columna 'pickup_longitude' con la mediana calculada.
df_taxi['pickup_longitude'].fillna(media_pickup_longitude, inplace=True)

print(f"La mediana de 'pickup_longitude' utilizada para rellenar es: {media_pickup_longitude:.6f}")
print(f"Valores nulos restantes en 'pickup_longitude': {df_taxi['pickup_longitude'].isnull().sum()}")
```

```python
# Calcula la mediana de la columna 'pickup_latitude'
media_pickup_latitude = df_taxi['pickup_latitude'].median()

# Rellena los valores nulos (NaN) en la columna 'pickup_latitude' con la mediana calculada.
df_taxi['pickup_latitude'].fillna(media_pickup_latitude, inplace=True)

print(f"La mediana de 'pickup_latitude' utilizada para rellenar es: {media_pickup_latitude:.6f}")
print(f"Valores nulos restantes en 'pickup_latitude': {df_taxi['pickup_latitude'].isnull().sum()}")
```

###cambio de tipo de datos

```python
#Cambiar tipos de datos de forma masiva usando un diccionario
# Usamos .astype() para números y pd.to_datetime() para fechas
df_taxi = df_taxi.astype({
    'vendor_id': 'category',
    'store_and_fwd_flag' : 'category',
    'payment_type' : 'category',
})

# Para fechas es mejor usar to_datetime para mayor precisión
# Asumiendo que 'pickup_datetime' y 'dropoff_datetime' son las columnas de fecha en df_taxi
df_taxi['pickup_datetime'] = pd.to_datetime(df_taxi['pickup_datetime'])
df_taxi['dropoff_datetime'] = pd.to_datetime(df_taxi['dropoff_datetime'])
```
